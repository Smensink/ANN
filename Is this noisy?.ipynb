```python
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# load json and create model
json_file = open('modelv5.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("modelv5.h5")
print("Loaded model from disk")
loaded_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
indir = '/Users/sebastianmensink/Downloads/CSV 2'
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        filename = os.path.join(indir, f)
        print (filename)
        dataset = pd.read_csv(filename)
        dataset= dataset.dropna(how = 'any')
        dataset = dataset.replace('_',0)
        X = dataset.iloc[:,0:12]
        X = sc.fit_transform(X)
        dataset['Prediction'] = loaded_model.predict(X)
        dataset.loc[dataset['Prediction'] > 0.5, 'NoiseP'] = 'Clean'
        dataset.loc[dataset['Prediction'] <= 0.5, 'NoiseP'] = 'Noisy'
        dataset.to_csv(filename, encoding='utf-8', index=False)
```
